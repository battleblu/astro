```python
import astropy.units as u
```

```python
import lightkurve as lk
```

```python
search_result = lk.search_lightcurve("8868650", author="Kepler", cadence="long")
search_result
```

```python
klc = lk.search_lightcurve("8868650", author="Kepler", cadence="long",quarter="13").download()
```

```python
klc
```

```python
ax = klc.plot()
```

```python
klc.scatter();
```

```python
pg=klc.normalize(unit='ppm').to_periodogram()
```

```python
pg.plot();
```

```python
pg.plot(scale='log',view='period');
```

```python
period=pg.period_at_max_power
print('Best period: {}'.format(period))
klc.fold(period.value).scatter();
```

```python
period=pg.period_at_max_power*4
print('Best period: {}'.format(period))
klc.fold(period.value).scatter();
```

```python
pg=klc.to_periodogram(minimum_period=4.44*u.day,maximum_period=8*u.day,oversample_factor=10000)
```

```python
period=pg.period_at_max_power
print('Best period: {}'.format(period))
klc.fold(period.value).scatter();
```
